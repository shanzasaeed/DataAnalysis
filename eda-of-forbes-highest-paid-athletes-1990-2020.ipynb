# Exploratory data analysis of Forbes Highest Paid Athletes 1990-2020 

Complete list of the world's highest-paid athletes since the first list published by Forbes in 1990. In 2002, they changed the reporting period from the full calendar year to June-to-June, and consequently, there are no records for 2001.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
athletes_df=pd.read_csv('/kaggle/input/forbes-highest-paid-athletes-19902019/Forbes Richest Atheletes (Forbes Richest Athletes 1990-2020).csv')#upload dataset
```

```python
athletes_df
```

## Top 10 highest paid Athletes (1990-2020)

```python
top_ten=athletes_df.sort_values('earnings ($ million)',ascending=False).head(10)
```

```python
top_ten
```

```python
plot=top_ten.plot.bar(x='Name',y='earnings ($ million)',figsize=(12,6));
plt.title('Top 10 highest paid Athletes (1990-2020)');
for bar in plot.patches:
  # ha and va stand for the horizontal and vertical alignment
    plot.annotate(format(bar.get_height(), '.1f'), (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha = 'center', va = 'center', 
                   size=8,xytext = (0, 8), 
                   textcoords = 'offset points')
plt.ylabel('Earnings ($ million)');
    
#bar graph of Top 10 highest paid athletes from 1990 till 2020, Athletes names on X axis and earning on Y axis. 
```

##  Sports in which top 10 athelets are 

```python
plot=top_ten.plot.bar(x='Sport',y='earnings ($ million)',figsize=(12,6));
plt.title('Top 10 highest paid Athletes (1990-2020)');
for bar in plot.patches:
  # ha and va stand for the horizontal and vertical alignment
    plot.annotate(format(bar.get_height(), '.1f'), (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha = 'center', va = 'center', 
                   size=8,xytext = (0, 8), 
                   textcoords = 'offset points')
plt.ylabel('Earnings ($ million)');
```

##  Decade wise top 10 highest paid Athletes:

```python
decade1=athletes_df[athletes_df.Year.isin(np.arange(1990,2000,1))]
```

```python
decade2=athletes_df[athletes_df.Year.isin(np.arange(2000,2011,1))]#because 2001 data is missing
```

```python
decade3=athletes_df[athletes_df.Year.isin(np.arange(2011,2021,1))]
```

```python
decade1=(decade1.sort_values('earnings ($ million)',ascending=False).head(10)).sort_values('Year')
decade2=(decade2.sort_values('earnings ($ million)',ascending=False).head(10)).sort_values('Year')
decade3=(decade3.sort_values('earnings ($ million)',ascending=False).head(10)).sort_values('Year')
```

```python
plots1=decade1.plot.bar(x='Year',y='earnings ($ million)',figsize=(10,5))
plt.title('Top 10 highest paid Athletes (1990-1999)')
plt.ylabel('Earnings ($ million)');
for bar in plots1.patches:
    plots1.annotate(format(bar.get_height(), '.1f'), (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha = 'center', va = 'center', 
                   size=8,xytext = (0, 8), 
                   textcoords = 'offset points')

plots2=decade2.plot.bar(x='Year',y='earnings ($ million)',figsize=(10,5))
plt.title('Top 10 highest paid Athletes (2000-2010)')
plt.ylabel('Earnings ($ million)');
for bar in plots2.patches:
    plots2.annotate(format(bar.get_height(), '.1f'), (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha = 'center', va = 'center', 
                   size=8,xytext = (0, 8), 
                   textcoords = 'offset points')

plots3=decade3.plot.bar(x='Year',y='earnings ($ million)',figsize=(10,5))
plt.title('Top 10 highest paid Athletes (2011-2020)')
plt.ylabel('Earnings ($ million)');
for bar in plots3.patches:
    plots3.annotate(format(bar.get_height(), '.1f'), (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha = 'center', va = 'center', 
                   size=8,xytext = (0, 8), 
                   textcoords = 'offset points')
```

## Countries which has most athletes in Forbes

```python
athletes_count_by_nationality=athletes_df.Nationality.value_counts().to_frame()
```

```python
athletes_count_by_nationality
```

```python
plt.figure(figsize=(30,27))
plt.pie(athletes_count_by_nationality.values.flatten(),
        labels=athletes_count_by_nationality.index.tolist(),
        autopct='%.1f%%')
plt.title('Athletes listed in Forbes by country (1990-2020)',fontsize=20)
plt.legend(athletes_count_by_nationality.index)
plt.show()
```

## Which sport has maximum number of athletes in Forbes, listed till 2020?

```python
no_of_athletes=athletes_df.Sport.str.lower().value_counts().to_frame()
#case sensitive, so first have to convert all upper case characters into lower case 
```

```python
no_of_athletes
```

```python
plt.figure(figsize=(30,27))
plt.pie(no_of_athletes.values.flatten(),
        labels=no_of_athletes.index.tolist(),
        autopct='%.1f%%')
plt.title('Number of athletes in each sport listed in Forbes (1990-2020)',fontsize=20)
plt.legend(no_of_athletes.index)
plt.show()
```

## Trend of average income of athletes 

```python
sport_df=athletes_df.groupby(athletes_df.Sport.str.lower())[['earnings ($ million)']].mean()
#group all athletes according to sport, and calculate mean of all athletes earnings, to see trend of most paid sports
```

```python
sport_df=sport_df.sort_values('earnings ($ million)')
```

```python
sport_df
```

```python
plt.figure(figsize=(15,6))

plt.plot(sport_df[['earnings ($ million)']].index[:].tolist(),sport_df[['earnings ($ million)']],marker='o')
plt.xticks(rotation=90)
plt.xlabel('Sport')
plt.ylabel('earnings ($ million)')
plt.title("Earning trend of sports whose athletes are mentioned in 'Forbes Highest Paid Athletes (1990-2000)' ")
plt.legend(sport_df[['earnings ($ million)']]);


#Earning analysis of each sport by taking average of earnings of all its athletes mentioned in 'Forbes Highest Paid Athletes (1990-2000)'.
```

## Year wise average earning of every sport

```python
athletes_df.Sport=athletes_df.Sport.str.lower()
```

```python
pivoted = athletes_df.pivot_table(index ='Year',
                         columns = 'Sport',
                         values= 'earnings ($ million)',
                         aggfunc = 'mean',
                         fill_value = 0)
#zero (0) means no athlete listed in Forbes for that particular sport in that year.
```

```python
plt.figure(figsize=(15,10))
plt.title('Year wise average income of every sport')
sns.heatmap(pivoted,annot= True, fmt='.1f');
```
